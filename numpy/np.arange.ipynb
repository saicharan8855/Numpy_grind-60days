### np.arange

```python
def np_arange(start_or_stop, stop=None, step=1, dtype=None):
    if stop is None:
        start, stop = 0, start_or_stop
    else:
        start = start_or_stop

    for name, val in [("start", start), ("stop", stop), ("step", step)]:
        if not isinstance(val, (int, float)):
            raise TypeError(
                f"'{name}' must be a numeric value, got {type(val).__name__}"
            )

    
    if step == 0:
        raise ValueError("np_arange: step cannot be zero")

   
    raw = (stop - start) / step
    n = int(raw) if raw == int(raw) else int(raw) + (1 if raw > 0 else 0)
    if n <= 0:
        result = []
    else:
        result = [start + i * step for i in range(n)]

   
    if dtype is not None:
        result = [dtype(val) for val in result]

    return result
```

### test cases :

```python
print(np_arange(1, 6))
print(np_arange(5, 0, -1) )
print(np_arange(0, 1, 0.25)) 
print(np_arange(-5, -1)) 
print(np_arange(5, 1))
print(np_arange(0, 1, 0.25, dtype=float))
```
