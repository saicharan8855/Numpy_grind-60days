### np.tan 

```python
import math

def get_shape(a):
    if not isinstance(a, list):
        return ()
    if len(a) == 0:
        return (0,)
    inner_shape = get_shape(a[0])
    for i in range(1, len(a)):
        if get_shape(a[i]) != inner_shape:
            raise ValueError(
                f"Jagged array: first row shape {inner_shape}, "
                f"but row {i} shape {get_shape(a[i])}"
            )
    return (len(a),) + inner_shape


def _build_nested(shape, getter, prefix=()):
    if len(shape) == 0:
        return getter(prefix)
    return [_build_nested(shape[1:], getter, prefix + (i,)) for i in range(shape[0])]


def _get_item(a, idx):
    cur = a
    for i in idx:
        cur = cur[i]
    return cur


def np_tan(a):
    shape = get_shape(a)
    if len(shape) == 0:
        return math.tan(a)

    def getter(idx):
        return math.tan(_get_item(a, idx))

    return _build_nested(shape, getter)
```

### test cases

```python
print(np_tan(0))
print(np_tan(math.pi / 4))
print(np_tan(math.pi / 3))
print(np_tan(math.pi / 6))
print(np_tan([0, math.pi / 6, math.pi / 4, math.pi / 3]))
print(np_tan([[0, math.pi / 4], [math.pi / 6, math.pi / 3]]))
print(np_tan([[0, math.pi / 4, math.pi / 3], [0, math.pi / 6, math.pi / 2 * 0.99]]))
```
