### np.linspace

```python
def np_linspace(start , stop , num = None):
    if num == 1: 
        return [start]
    if num == 0: 
        return []
    if num is None:
        num = 50
    res = []
    step = (stop - start) / (num - 1)
    for i in range(0 , num  ):
        val = start + i * step 
        res.append(val)

    return [res]
```

```python
print(np_linspace(0,10))
```
