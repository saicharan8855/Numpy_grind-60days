### np.log2

```python
import math

def get_shape(a):
    if not isinstance(a, list):
        return ()
    if len(a) == 0:
        return (0,)
    inner_shape = get_shape(a[0])
    for i in range(1, len(a)):
        if get_shape(a[i]) != inner_shape:
            raise ValueError(
                f"Jagged array: first row shape {inner_shape}, "
                f"but row {i} shape {get_shape(a[i])}"
            )
    return (len(a),) + inner_shape


def _build_nested(shape, getter, prefix=()):
    if len(shape) == 0:
        return getter(prefix)
    return [_build_nested(shape[1:], getter, prefix + (i,)) for i in range(shape[0])]


def _get_item(a, idx):
    cur = a
    for i in idx:
        cur = cur[i]
    return cur


def _log2_scalar(x):
    if x > 0:
        return math.log2(x)
    if x == 0:
        return float('-inf')
    return float('nan')


def np_log2(a):
    shape = get_shape(a)
    if len(shape) == 0:
        return _log2_scalar(a)

    def getter(idx):
        return _log2_scalar(_get_item(a, idx))

    return _build_nested(shape, getter)
```

### test cases 

```python
print(np_log2(1))
print(np_log2(2))
print(np_log2(4))
print(np_log2(8))
print(np_log2([1, 2, 4, 8, 16]))
print(np_log2([[1, 2], [4, 8]]))
print(np_log2([[[1, 2], [4, 8]], [[16, 32], [64, 128]]]))
print(np_log2(0))
print(np_log2(-1))
print(np_log2([-1, 0, 1, 2]))
```
