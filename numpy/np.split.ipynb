### np.split

```python
def get_ndim(a):
    if not isinstance(a , list):
        return 0
    if len(a) == 0: 
        return 1
    return 1 + get_ndim(a[0])

def axis_length(a , axis):
    if axis == 0:
        return len(a)
    if len(a) == 0: 
        return 0
    return axis_length(a[0] , axis - 1)

def _split_axis0(a, cuts):

    n = len(a) 
    cuts = [c for c in cuts if  0 <= c <= n]
    cuts =  sorted(cuts)
    bounds = [0] + cuts + [n]
    result = []
    for i in range(len(bounds) - 1): 
        start = bounds[i]
        stop = bounds[i + 1]
        result.append(a[start:stop])
    return result

def _split_recursive(a, cuts, axis):
    if axis == 0:
        return _split_axis0(a, cuts)

    per_row = [_split_recursive(row, cuts, axis - 1) for row in a]
    num_parts = len(per_row[0])
    result = []
    for k in range(num_parts):
        part_k = [row_parts[k] for row_parts in per_row]
        result.append(part_k)
    return result


def np_split(array , indices_or_section , axis = 0): 
    if not isinstance(array , list):
        raise TypeError("input must be type of list")
    ndim = get_ndim(array) 
    if axis < 0: 
        axis = ndim + axis
    if not 0 <= axis < ndim:
        raise ValueError("validate axis")
    length = axis_length(array , axis)
    if isinstance(indices_or_section , int): 
        N = indices_or_section
        if N <= 0: 
            raise ValueError("length of indices should be positve ")
        if length % N != 0: 
            raise ValueError("array spplit will not result equals")
        step = length // N 
        cuts = [step * k for k in range(1 , N)]
    else:
        cuts = list(indices_or_section)
    return _split_recursive(array, cuts, axis)
    
```
