### np.sum

```python
def get_ndim(a):
    if not isinstance(a, list):
        return 0
    if len(a) == 0:     
        return 1
    return 1 + get_ndim(a[0])

def np_sum(a , initial = 0 , axis = None ):
    ndim = get_ndim(a)
    if not isinstance(initial , (int , float)):
        raise TypeError("initial should be integer or float")
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input should be type of int or float")
        return a
    elif isinstance(a , list):
        if axis is not None and axis >= ndim:
            raise ValueError("axis out of range")
        if axis is not None and axis < 0:
            axis = ndim + axis
        if all(isinstance(row, list) for row in a): 
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must have equal length")


        if axis is None:
            total = initial
            for i in range(len(a)):
                total += np_sum(a[i])
            return total

        elif axis == 1:
            result = []
            for i in range(len(a)):
                result.append(np_sum(a[i]))
            return result

        elif axis == 0:
            result = []
            for j in range(len(a[0])):
                total = 0
                for i in range(len(a)):
                    total += np_sum(a[i][j])  
                result.append(total)
            return result  
            
```
