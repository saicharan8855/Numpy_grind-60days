### np.median

```python
def get_ndim(a):
    if not isinstance(a , list):
        return 0
    if len(a) == 0:
        return 1
    return 1 + get_ndim(a[0])
def validate_numeric(a):
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input must be of type int or float")
        return a 
    if isinstance(a , list):
        for i in range(len(a)):
            validate_numeric(a[i])

def np_median(a ,  axis =  None ):
    ndim = get_ndim(a)
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input must be type of int or float")
        return a
    else:
        validate_numeric(a)
        if len(a) == 0: 
            raise ValueError("input should not be empty")
            
        if axis is not None and axis >= ndim:
            raise ValueError("ndim should be greater than axis")
        if axis  is not None and axis < 0:
            axis = ndim + axis
        if all(isinstance(row, list) for row in a): 
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must have equal length")
            
        if axis is None:
            flat = sorted(a)
            n = len(a)
            if n % 2 == 0:
                return (flat[n//2 - 1 ] + flat[n//2])/2

            if n % 2 != 0:
                return flat[n//2]

        if axis == 1:
            result = []
            for i in range(len(a)):
                result.append(np_median(a[i]))
            return result
        if axis == 0:
            if not isinstance(a[0], list):
                return np_median(a)
            result = []
            for j in range(len(a[0])):
                col = [a[i][j] for i in range(len(a))]
                result.append(np_median(col))
            return result
```
