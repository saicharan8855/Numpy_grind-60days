### np.floor

```python
def np_floor(a):
    if not isinstance(a , list):
        if isinstance(a,complex):
            raise TypeError("complex numbers arent allowed")
        if not isinstance(a , (int , float)):
            raise TypeError("input must be type of int or float")
        if a < 0:
            frac = a - int(a)
            return  int(a) if frac == 0 else int(a) - 1    
        if a > 0:
            return int(a)
        

    # empty list
    if isinstance(a,list):
        if len(a) == 0:
            return [] 

        # list
        result = []
        for i in range(len(a)):
            result.append(np_floor(a[i]))
        return result
```

### test_cases :

```python
print(np_floor(3.7))                   
print(np_floor(-3.7))                  
print(np_floor(3.0))                 
print(np_floor(-3.0))                  
print(np_floor([3.7, -3.7, 3.0]))      
print(np_floor([[3.7,-3.7],[1.1,-1.1]]))  
print(np_floor([]))                    
        
```
