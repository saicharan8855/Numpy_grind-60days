### np.round

```python
def np_round(a, decimals=0):
    if not isinstance(a, list):
        if isinstance(a, complex):
            raise TypeError("complex numbers aren't allowed")
        if not isinstance(a, (int, float)):
            raise TypeError("input must be type of int or float")
        factor = 10 ** decimals
        n = (a * factor) + 0.5
        if n < 0:
            frac = n - int(n)
            result = int(n) if frac == 0 else int(n) - 1
        else:
            result = int(n)
        return result / factor

    if isinstance(a, list):
        if len(a) == 0:
            return []
        result = []
        for i in range(len(a)):
            result.append(np_round(a[i], decimals))
        return result
```

### test case :

```python
print(np_round(3.7))                      
print(np_round(3.2))                      
print(np_round(3.5))                      
print(np_round(-3.5))                     
print(np_round(3.456, 2))                 
print(np_round([3.2, 3.7, 3.5]))          
print(np_round([[3.2, 3.7],[3.5, -3.5]])) 
print(np_round([3.456, 2.789], 2))        
print(np_round([]))                       
                      
```
