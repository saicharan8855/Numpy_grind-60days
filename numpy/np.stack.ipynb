### np.stack 

```python
def get_shape(a):
    if not isinstance(a , list): 
        return ()
    if len(a) == 0: 
        return (0, )
    inner_shape = get_shape(a[0])

    for i in range(1 , len(a)): 
        if  get_shape(a[i]) != inner_shape:
            raise ValueError("jagged array detected")
    return (len(a), ) + inner_shape


def get_ndim(a):
    if not isinstance(a, list):
        return 0
    if len(a) == 0:
        return 1
    return 1 + get_ndim(a[0])


def np_stack(arrays , axis = 0):
    if len(arrays) == 0: 
        raise ValueError("inputs should not be empty")
    shape = get_shape(arrays[0])
    ndim = get_ndim(arrays[0])

    if axis < 0:
        axis = (ndim + 1) + axis 
    if not (0 <= axis <= ndim):
        raise ValueError("Validate axis")
    for arr in arrays[1:]: 
        if not get_shape(arr) == shape:
            raise ValueError("all arrays must have same shape") 

    if axis == 0:
        return list(arrays)
    if axis > 0:
        result = []
        for i in range(len(arrays[0])): 
            group = [arr[i] for arr in arrays]
            result.append(np_stack(group , axis - 1))
        return result
```
