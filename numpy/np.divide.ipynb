### np.divide

```python
def np_divide(a , b):
    # if both inputs are scalars
    if b == 0:
        raise ZeroDivisionError("cant divide by zero")
    if not isinstance(a , list) and not isinstance(b , list):
        if not isinstance(a , (int , float)) or not isinstance(b ,(int , float)):
            raise TypeError("inputs must be type of int or float")
        if b == 0:
            raise ZeroDivisionError("cant divide by zero")
        return a / b

    # if one is list and another is scaler

    # broadcasting
    if isinstance(a , list) and not isinstance(b , list):
        result = []
        for x in a:
            result.append(np_divide(x , b))
        return result

    if isinstance(b,list) and not isinstance(a , list):
        result = []
        for y in b:
            result.append(np_divide(y , a))
        return result

    # if both are lists
    # empty list
    if isinstance(a,list) and isinstance(b , list):
        if len(a) == 0 and len(b) == 0:
            return []
        if len(a) == 0 or len(b) == 0:
            raise ValueError("inputs shouldnt be non - empty")

        if len(a) != len(b):
            raise ValueError("length of inputs should be same")

        result = []
        for i in range(len(a)):
            result.append(np_divide(a[i],b[i]))
        return result
            
```
