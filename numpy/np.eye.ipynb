### np.eye

```python
def np_eye(N , M = None, k = 0 , dtype = float):
    if M is None:
        M = N
    if not isinstance(N , int) or not isinstance(M , int):
        raise TypeError("inputs N and M must be type of int")
    if N < 0 or M < 0: 
        raise ValueError("N and M must be non negative")
    if not isinstance(k , int):
        raise TypeError(" k should be an integer")
    if not callable(dtype):
        raise TypeError("dtype should be callable")
        

    result = []
    for i in range(N):
        one_row = []
        for j in range(M):
            if j == i + k:
                one_row.append(dtype(1))
            else:
                one_row.append(dtype(0))
        result.append(one_row)
    return result
```
