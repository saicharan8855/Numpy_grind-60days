### np.max

```python
def get_ndim(a):
    if not isinstance(a, list):
        return 0
    if len(a) == 0:     
        return 1
    return 1 + get_ndim(a[0])

    
def np_max(a,initial = 0 , axis = None):
    ndim = get_ndim(a)
    if not isinstance(initial , (int , float)):
        raise TypeError("initial should be integer or float")
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input should be type of int or float")
        return a
    elif isinstance(a , list):
        if axis is not None and axis >= ndim:
            raise ValueError("axis out of range")
        if axis is not None and axis < 0:
            axis = ndim + axis
        if all(isinstance(row, list) for row in a): 
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must have equal length")
        if axis is None:
            maxi = np_max(a[0])
            for i in range(1 , len(a)):
                val = np_max(a[i])
                if val < maxi:
                    maxi = val
            return maxi

        elif axis == 1:
            result = []
            for i in range(len(a)):
                result.append(np_max(a[i]))
            return result

        elif axis == 0:
            if not isinstance(a[0], list):  
                return np_max(a)
            result = []
            for j in range(len(a[0])):
                col = [a[i][j] for i in range(len(a))]
                result.append(np_max(col))  
            return result 
```

```python
print(np_max([3, 1, 2]))                          
print(np_max([-3, -1, -7]))                        
print(np_max([[3,1],[4,0]], axis=1))               
print(np_max([[3,1],[4,0]], axis=0))               
print(np_max([[[1,2],[3,4]],[[5,6],[7,8]]]))       
print(np_max([1, 2, 3], axis=-1))                  
print(np_max([[3,1],[4,0]], axis=-1))               
print(np_max([[3,1],[4,0]], axis=-2))              
```
