### np.ones

```python
def np_ones(a , dtype = float):
    if not isinstance(a,tuple):
        raise TypeError("input must be of type tuple")
    for dim in a:
        if not isinstance(dim , int):
            raise TypeError(f"{dim} is not an integer")
        if dim < 0: 
            raise ValueError(f"{dim} is negative")
    if not callable(dtype):
        raise TypeError(f"dtype must be callable but got {type(dtype).__name__}")
    one = dtype(1)
    if len(a) == 0: 
        return one
    if len(a) == 1: 
        return [one] * a[0]
    else:
        result = []
        for i in range(a[0]):
            result.append(np_ones(a[1:] , dtype))
        return result
```

### test cases : 

```python
print(np_ones((3,)))                  
print(np_ones((2, 2, 3)))             
print(np_ones((3,), dtype=int))       
print(np_ones((2, 2), dtype=bool))    
print(np_ones(()))                    
print(np_ones((0,)))                  
print(np_ones((2, 0, 3)))      
```
