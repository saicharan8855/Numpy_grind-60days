### np.arcsin

```python
import math

def get_shape(a):
    if not isinstance(a, list):
        return ()
    if len(a) == 0:
        return (0,)
    inner_shape = get_shape(a[0])
    for i in range(1, len(a)):
        if get_shape(a[i]) != inner_shape:
            raise ValueError(
                f"Jagged array: first row shape {inner_shape}, "
                f"but row {i} shape {get_shape(a[i])}"
            )
    return (len(a),) + inner_shape


def _build_nested(shape, getter, prefix=()):
    if len(shape) == 0:
        return getter(prefix)
    return [_build_nested(shape[1:], getter, prefix + (i,)) for i in range(shape[0])]


def _get_item(a, idx):
    cur = a
    for i in idx:
        cur = cur[i]
    return cur


def _arcsin_scalar(x):
    if x < -1 or x > 1:
        return float('nan')
    return math.asin(x)


def np_arcsin(a):
    shape = get_shape(a)
    if len(shape) == 0:
        return _arcsin_scalar(a)

    def getter(idx):
        value = _get_item(a, idx)
        return _arcsin_scalar(value)

    return _build_nested(shape, getter)
```

### test cases 

```python
print(np_arcsin(-1))
print(np_arcsin(-0.5))
print(np_arcsin(0))
print(np_arcsin(0.5))
print(np_arcsin(1))
print(np_arcsin([0, 0.5, 1, -0.5, -1]))
print(np_arcsin([0, 1, 0.3, -1]))
print(np_arcsin([[0, 0.5], [1, -0.5]]))
print(np_arcsin([[0, 1], [0.5, -1]]))
print(np_arcsin([-2, -1.5, 0, 1.5, 2]))
print(np_arcsin([]))
print(np_arcsin([[]]))
```
