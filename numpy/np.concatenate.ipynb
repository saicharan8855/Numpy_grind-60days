### np.concatenate

```python
def get_shape(a):
    if not isinstance( a, list):
        return ()
    if len(a) == 0:
        return (0 ,)
    inner_shape = get_shape(a[0])

    for i in range(1 , len(a)):
        if get_shape(a[i]) != inner_shape:
            raise ValueError("jagged array detected")

    return (len(a), )  + inner_shape

def get_ndim(a): 
    if not isinstance(a , list): 
        return 0
    if len(a) == 0: 
        return 1
    return 1 + get_ndim(a[0])

def np_flatten(a):
    if not isinstance( a, list):
        return [a]
    result = []
    for item in a:
        result.extend(np_flatten(item))
    return result
        

def np_concatenate(arrays , axis = 0): 
    if len(arrays) == 0: 
        raise ValueError("empty array is not allowed")
    if not isinstance(arrays , (list , tuple)):
        raise TypeError(f"input must be list or tuple but got {type(arrays).__name__}")

    if axis is None:
        result = []
        for arr in arrays:
            result.extend(np_flatten(arr))
        return result

    ndim = get_ndim(arrays[0])
    if axis < 0: 
        axis = ndim + axis
    if not (0 <= axis < get_ndim(arrays[0])): 
        raise ValueError("Out of bounds")

    if axis == 0: 
        result = []
        for arr in arrays: 
            result.extend(arr)
        return result

    if axis > 0: 
        result = [] 
        for i in range(len(arrays[0])):
            group = [arr[i] for arr in arrays]
            merged = np_concatenate(group , axis - 1)
            result.append(merged)
        return result
```
