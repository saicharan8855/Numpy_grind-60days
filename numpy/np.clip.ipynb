### np.clip

```python
def np_clip(x , x_min , x_max):
    if x_min > x_max :
        raise ValueError("min should be less than maximum")

    # for lists
    if isinstance(x , list):
        if len(x) == 0 :
            return []

        result = []
        for i in range(len(x)) :
            result.append(np_clip(x[i] , x_min , x_max ))
        return result

    # for scalar
    if not isinstance(x,(int , float)):
        raise ValueError("input must be int or float")
    if x < x_min:
        return x_min
    if x > x_max:
        return x_max
    return x    
```

### test_cases :

```python
print(np_clip(5, 1, 3))                        
print(np_clip(0, 1, 3))                        
print(np_clip(2, 1, 3))                        
print(np_clip([0,2,5], 1, 3))                  
print(np_clip([[0,2],[5,1]], 1, 3))            
print(np_clip([], 1, 3))                       
```
