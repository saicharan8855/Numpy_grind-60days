### np.mean , np.size

```python
def np_size(a):
    if not isinstance(a , list):
        return 1
    res = []
    for i  in range(len(a)):
        res.append(np_size(a[i]))
    return sum(res)
        
```

```python
print(np_size([1,2,3]))        
print(np_size([[1,2],[3,4]]))   
print(np_size(5))              
```

```python
def get_ndim(a):
    if not isinstance(a, list):
        return 0
    if len(a) == 0:     
        return 1
    return 1 + get_ndim(a[0])

def np_sum(a , initial = 0 , axis = None ):
    ndim = get_ndim(a)
    if not isinstance(initial , (int , float)):
        raise TypeError("initial should be integer or float")
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input should be type of int or float")
        return a
    elif isinstance(a , list):
        if axis is not None and axis >= ndim:
            raise ValueError("axis out of range")
        if axis is not None and axis < 0:
            axis = ndim + axis
        if all(isinstance(row, list) for row in a): 
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must have equal length")


        if axis is None:
            total = initial
            for i in range(len(a)):
                total += np_sum(a[i])
            return total

        elif axis == 1:
            result = []
            for i in range(len(a)):
                result.append(np_sum(a[i]))
            return result

        elif axis == 0:
            result = []
            for j in range(len(a[0])):
                total = 0
                for i in range(len(a)):
                    total += np_sum(a[i][j])  
                result.append(total)
            return result
```

```python
def validate_numeric(a):
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input must be of type int or float")
        return a 
    if isinstance(a , list):
        for i in range(len(a)):
            validate_numeric(a[i])
            
                
```

```python
def np_mean(a , axis = None):
    ndim = get_ndim(a)
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("inputs must be type of int or float")
        return a
    elif isinstance(a , list):
        validate_numeric(a)
        if np_size(a) == 0:
            raise ValueError("input list is empty")
        if axis is not None and axis >= ndim:
            raise ValueError("axis should be less than ndim")
        if axis is not None and axis < 0:
            axis = ndim + axis
        if all(isinstance(row, list) for row in a): 
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must have equal length")
        if axis == None:
            mean = np_sum(a)/np_size(a)
            return mean
        if axis == 1:
            res = []
            for i in range(len(a)):
                res.append(np_mean(a[i]))
            return res
        if axis == 0:
            
            if not isinstance(a[0], list):
                return np_mean(a)
            res= []
            for j in range(len(a[0])):
                col = [a[i][j] for i in range(len(a)) ]
                res.append(np_mean(col))
            return res
            
            
```
