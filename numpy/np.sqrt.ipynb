### np.sqrt

```python
def np_sqrt(a):
    
    if not isinstance(a , list):
        if not isinstance(a , (int , float)):
            raise TypeError("input must be type of int or float")
        if a < 0:
            raise ValueError("input must be a positive number")
        return a ** 0.5

    # empty list
    if isinstance(a,list):
        if len(a) == 0:
            return []
        

        result = []
        for i in range(len(a)):
            result.append(np_sqrt(a[i]))
        return result
            
```

### testcases :

```python
print(np_sqrt(9))
print(np_sqrt(2))                   
print(np_sqrt([4, 9, 16]))           
print(np_sqrt([[4,9],[16,25]]))      
print(np_sqrt([]))                   
            
```
