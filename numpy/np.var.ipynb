## np.var

```python
def ndim(a):
    if not isinstance(a , list):
        return 0
    if len(a) == 0:
        return 1
    return 1 + ndim(a[0])

def validate_numeric(a):
    if not isinstance(a , list):
        if not isinstance(a , (int , float)): 
            raise ValueError("input must be of type int or float")
        return a 
    if isinstance(a , list): 
       for i in range(len(a)):
           validate_numeric(a[i])

def np_var(a , axis = None):
    n_dim = ndim(a)
    if not isinstance( a , list):
        if not isinstance(a , (int , float)):
            raise ValueError("input must be type of int or float")
        return a
    elif isinstance(a , list):
        if len(a) == 0:
            raise ValueError("len of list should not be empty")
            
        validate_numeric(a)
        if axis is not None and axis >= n_dim:
            raise ValueError("axis should be greater than ndim")
        if axis is not None and axis < 0:
            axis = n_dim + axis
        if all(isinstance(row , list) for row in a):
            for row in a:
                if len(row) != len(a[0]):
                    raise ValueError("all rows must should be of same length")
        
        if axis is None:
            total = 0
            for i in range(len(a)):
                total += a[i]
            mean = total / len(a)
            sum_sq = 0 
            for j in range(len(a)):
                sum_sq += ((a[j] - mean)**2)
            variance = sum_sq/len(a)
            return variance
            
                
                
        if axis == 1:
            res = []
            for i in range(len(a)):
                res.append(np_var(a[i]))
            return res

        if axis == 0:
            if not isinstance(a[0] , list):
                return np_var(a)
            res= []
            for j in range(len(a[0])):
                column = [a[i][j] for i in range(len(a))]
                res.append(np_var(column))
            return res
    
```
