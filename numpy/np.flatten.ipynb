### np.flatten

```python
def np_flatten(a):
    result = []
    if not isinstance( a , list):
        return [a]
    for item in a:
        result.extend(np_flatten(item))
    return result
```

### test cases :

```python
print(np_flatten([[1, 2], [3, 4]]))              
print(np_flatten([[[0,1],[2,3]], [[4,5],[6,7]]]))   
print(np_flatten(5))                                  
```

```python
np_flatten([[1,2],[3,4]], order='F')
```
