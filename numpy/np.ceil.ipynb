### np.ceil

```python
def np_ceil(a):
    if not isinstance(a , list):
        if isinstance(a,complex):
            raise TypeError("complex numbers arent allowed")
        if not isinstance(a , (int , float)):
            raise TypeError("input must be type of int or float")
        if a < 0:
            return  int(a)   
        if a > 0:
            frac = a - int(a) 
            return int(a) if frac == 0 else int(a) + 1
        

    # empty list
    if isinstance(a,list):
        if len(a) == 0:
            return [] 

        # list
        result = []
        for i in range(len(a)):
            result.append(np_ceil(a[i]))
        return result
```

### test_cases :

```python
print(np_ceil(3.2))                   
print(np_ceil(-3.7))                   
print(np_ceil(3.0))                    
print(np_ceil(-3.0))                   
print(np_ceil([3.2, -3.7, 3.0]))       
print(np_ceil([[3.2,-3.7],[1.1,-1.1]]))  
print(np_ceil([]))                     
                   
```
