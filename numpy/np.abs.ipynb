### np.abs

```python
def np_abs(a):
    if not isinstance(a , list):
        if isinstance(a,complex):
            return (a.real ** 2 + a.imag ** 2) ** 0.5
        if not isinstance(a , (int , float)):
            raise TypeError("input must be type of int or float")
        if a < 0:
            return -a    # -(-a) = a 
        return a

    # empty list
    if isinstance(a,list):
        if len(a) == 0:
            return [] 

        # list
        result = []
        for i in range(len(a)):
            result.append(np_abs(a[i]))
        return result
```

### test cases :

```python
print(np_abs(-5))                   
print(np_abs(3.5))                   
print(np_abs(3+4j))                 
print(np_abs([-1,-2,-3]))           
print(np_abs([[- 1,-2],[3,-4]]))     
print(np_abs([3+4j, -5]))            
print(np_abs([]))            
```
