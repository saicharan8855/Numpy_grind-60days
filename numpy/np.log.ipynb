### np.log

```python
import math

def get_shape(a):
    if not isinstance(a, list):
        return ()
    if len(a) == 0:
        return (0,)
    inner_shape = get_shape(a[0])
    for i in range(1, len(a)):
        if get_shape(a[i]) != inner_shape:
            raise ValueError(
                f"Jagged array: first row shape {inner_shape}, "
                f"but row {i} shape {get_shape(a[i])}"
            )
    return (len(a),) + inner_shape


def _build_nested(shape, getter, prefix=()):
    if len(shape) == 0:
        return getter(prefix)
    return [_build_nested(shape[1:], getter, prefix + (i,)) for i in range(shape[0])]


def _get_item(a, idx):
    cur = a
    for i in idx:
        cur = cur[i]
    return cur


def np_log(a):
    shape = get_shape(a)
    if len(shape) == 0:
        return math.log(a)

    def getter(idx):
        return math.log(_get_item(a, idx))

    return _build_nested(shape, getter)
```

### testcasess 

```python
print(np_log(1))
print(np_log(math.e))
print(np_log(10))
print(np_log([1, math.e, math.e**2, 100]))
print(np_log([[1, math.e], [math.e**3, 100]]))
print(np_log([[[1, math.e], [math.e**2, 100]], [[10, 100], [1000, 10000]]]))
print(np_log([[]]))
print(np_log([]))
```
